```python
'''
Complete the repeatedString function in the editor below. It should return an integer representing the number of occurrences of a in the prefix of length  in the infinitely repeating string.

repeatedString has the following parameter(s):

s: a string to repeat
n: the number of characters to consider


The first line contains a single string, . 
The second line contains an integer, .


Input
    
    aba
    10
    
Output

    7
'''
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the repeatedString function below.
def repeatedString(s, n):
    
    c = 'a'     # char to evalue
    lS = len(s) # len of string
    rep = 0     # number of repeated Char
    
    if (lS < n):
        # Case number of repearences are bigger than len of string given
        dRt = int(n / lS)    # Division result
        dRs = int(n % lS)    # Division resiude 
        
        coun = 0        # Counter
        
        for i in s:
            if i == c :
                coun += 1
        
        rep += (coun*dRt)
        coun = 0
        
        for i in range(dRs):
            if s[i] == c :
                coun += 1
        
        rep += coun
        
        return rep
        
    elif (lS > n):
        # Case len of string given is bigger than repearences
        for i in range(n):
            if s[i] == c :
                rep += 1
    
        
    elif (lS == n):
        # Case are equal
        for i in s:
            if i == c :
                rep += 1
    
    return rep
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    n = int(input())

    result = repeatedString(s, n)

    fptr.write(str(result) + '\n')

    fptr.close()
```
